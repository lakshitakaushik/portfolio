```python
grouped_df = df.groupby('Company')
```

```python
def kurtosis_value(series):
  return kurtosis(series)

def skewness_value(series):
  return skew(series)
descript=grouped_df['Open Price'].agg(['count','mean','std','min','max','var'])
print(descript)
```
